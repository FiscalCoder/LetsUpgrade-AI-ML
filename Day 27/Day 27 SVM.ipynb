## Importing modules

```python
import pandas as pd
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn import neighbors
from sklearn import svm
from sklearn.neighbors import KNeighborsClassifier
```

```python
dataset = pd.read_csv("train.csv")
dataset.columns
```

## Data Processing

```python
dataset.isna().sum()
dataset = dataset.drop(["Ticket", "Cabin", "PassengerId", "Name"], axis=1)
dataset.columns
```

```python
le = preprocessing.LabelEncoder()
le.fit(dataset["Sex"])
print(le.classes_)
dataset["Sex"] = le.transform(dataset["Sex"])
dataset["Embarked"] = le.fit_transform(dataset["Embarked"])
dataset.head()
```

## Support Vector Machine

```python
clf = svm.SVC(gamma=0.01,C=100)

def SVM(x,y):
    x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=0)
    clf.fit(x_train,y_train)
    y_pred = clf.predict(x_test)
    print(accuracy_score(y_test, y_pred, normalize=True))
```

```python
columns = ['Survived', 'Pclass', 'Sex', 'SibSp', 'Parch','Embarked']

c1 = columns[:]
 
for x in c1:
    w = dataset[x]
    q = dataset.drop([x], axis=1)
    print("When the dependent variable is", x)
    SVM(q,w)
    print()
```

### When the dependent varibale is Pclass the model accuracy increases exponentially. Hence Pclass should be taken as the dependent variable for SVM in the above dataset

## K_Nearest Neighbour

```python
def prediction(n):
    x = dataset.drop(['Pclass'],axis=1)
    y = dataset['Pclass']
    k = []
    s = []
    
    for i in range(1,n):
        knn = KNeighborsClassifier(n_neighbors = i)
        k.append(i)
        
        print()
    
        x_train, x_test, y_train, y_test = train_test_split(x, y, test_size =0.3, random_state=0)
        model = knn.fit(x_train,y_train)
        pred = model.predict(x_test)
        
        score = model.score(x_test,y_test)
        s.append(score)
    
        print(f"\nAcuuracy when k is {i} ", score)
```

```python
prediction(268)
```
